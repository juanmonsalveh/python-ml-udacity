```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('./data/fuel_econ.csv')
fuel_econ.head()
```

```python
# YOUR CODE HERE
# sb.regplot(data = fuel_econ, x = 'city', y = 'highway', x_jitter=0.04, scatter_kws = {'alpha' : 1/10}, fit_reg=False);

sb.regplot(data = fuel_econ, x = 'city', y = 'highway', scatter_kws = {'alpha' : 1/40});
plt.xlabel('Fuel mileage ratings for city')
plt.ylabel('Highway driving');
```
