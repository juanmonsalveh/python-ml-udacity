```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('./data/fuel_econ.csv')
print(fuel_econ.shape)
fuel_econ.head(6)
```

```python
plt.figure(figsize = [15, 5]) # larger figure size for subplots

plt.subplot(1, 2, 1) # 1 row, 2 cols, subplot 1
plt.scatter(data = fuel_econ, x = 'displ', y = 'comb');
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');

plt.subplot(1, 2, 2) # 1 row, 2 cols, subplot 2
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb', fit_reg=True);
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

```python
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter=0.3,
          scatter_kws = {'alpha' : 1/20});
plt.xlabel('Year')
plt.ylabel('Combined Fuel Eff. (mpg)');

# plt.scatter(data = fuel_econ, x = 'year', y = 'comb', alpha = 1/5, );
# plt.xlabel('Year')
# plt.ylabel('Combined Fuel Eff. (mpg)');
```
